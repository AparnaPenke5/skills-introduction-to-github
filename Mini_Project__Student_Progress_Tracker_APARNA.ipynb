```

# STUDENT PROGRESS TRACKER SYSTEM

while True:
    print("STUDENT PROGRESS TRACKER SYSTEM")

    name = input("Enter Student Name: ")
    student_class = input("Enter Class: ")
    roll = input("Enter Roll Number: ")

    subjects = int(input("\nHow many subjects do you want to enter? "))

    total = 0
    fail_flag = False  # If any subject < 35, student fails

    for i in range(1, subjects + 1):
        mark = float(input(f"Enter marks for subject {i}: "))
        total += mark
        if mark < 35:
            fail_flag = True  # at least one failed subject

    average = total / subjects

    if average >= 90:
        grade = "A"
    elif average >= 75:
        grade = "B"
    elif average >= 50:
        grade = "C"
    elif average >= 35:
        grade = "D"
    else:
        grade = "Fail"

    if fail_flag or grade == "Fail":
        result = "FAIL"
    else:
        result = "PASS"

    print("        STUDENT PROGRESS REPORT")
    print(f"Name         : {name}")
    print(f"Class        : {student_class}")
    print(f"Roll Number  : {roll}")
    print(f"Total Marks  : {total}")
    print(f"Average      : {average:.2f}")
    print(f"Grade        : {grade}")
    print(f"Result       : {result}")

    while True:  # Menu loop
        print("\nWhat do you want to do next?")
        print("1. Enter marks again")
        print("2. View report again")
        print("3. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            break  # Restart whole system (outer loop)
        elif choice == "2":


            print("        STUDENT PROGRESS REPORT")

            print(f"Name         : {name}")
            print(f"Class        : {student_class}")
            print(f"Roll Number  : {roll}")
            print(f"Total Marks  : {total}")
            print(f"Average      : {average:.2f}")
            print(f"Grade        : {grade}")
            print(f"Result       : {result}")

        elif choice == "3":
            print("\nExiting Program... Goodbye!")
            exit()  # End everything immediately
        else:
            print("Invalid choice! Try again.")
```
